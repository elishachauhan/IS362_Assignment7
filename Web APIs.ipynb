##### Import library modules: 

```python
import pandas as pd
import json
import urllib.request
from pandas.io.json import json_normalize
```

##### Build interface that reads the New York Times API that contains JSON data for the most viewed technology articles:

```python
api_key = '1bd53b115d014e1eab9c54a8c195d0b8'
url = 'https://api.nytimes.com/svc/mostpopular/v2/mostviewed/Technology/30.json?'
string = url+'&api-key='+api_key

response = str(urllib.request.urlopen(string).read(), 'utf-8')
```

##### Read JSON data onto pandas DataFrame:

```python
df = pd.read_json(response)
df
```

##### Load most viewed technology articles JSON data into pandas DataFrame that is sorted by views and shown in descending order: 

```python
data = json.loads(response)
most_viewed = json_normalize(data['results'])
most_viewed.sort_values(by=["views"], ascending=False)
```
